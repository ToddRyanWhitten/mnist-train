# MNIST dataset

Step 1: Acquire and train the MNIST dataset

> This script utilizes TensorFlow and Keras to build, train, and evaluate a convolutional neural network for handwritten digit classification using the MNIST dataset. It loads the dataset, normalizes the pixel values, defines a sequential model with dense layers and dropout, compiles it with the Adam optimizer and sparse categorical crossentropy loss, trains the model for 5 epochs, evaluates its performance on the test set, and finally saves the trained model as 'mnist.keras' in a 'model' directory.

```python
import os
import tensorflow as tf

def train_mnist():
    print("Creating trained model from mnist dataset sourced from keras")
    print("TensorFlow version:", tf.__version__)

    # Load and prepare the MNIST dataset
    mnist = tf.keras.datasets.mnist
    (x_train, y_train), (x_test, y_test) = mnist.load_data()

    # Normalize pixel values
    x_train, x_test = x_train / 255.0, x_test / 255.0

    # Build the Sequential model
    model = tf.keras.models.Sequential([
        tf.keras.layers.Flatten(input_shape=(28, 28)),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(10, activation='softmax')
    ])

    # Compile the model
    # Note: sparse_categorical_crossentropy is used because the labels are integers (0-9)
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    # Train the model
    print("Training model...")
    model.fit(x_train, y_train, epochs=5)

    # Evaluate the model
    print("\nEvaluating model...")
    model.evaluate(x_test, y_test, verbose=2)

    # Save the model in HDF5 format for compatibility with tensorflowjs_converter
    print("\nSaving model to mnist.h5...")
    os.makedirs("model", exist_ok=True)
    model.save("model/mnist.h5", save_format="h5")
```

```python
train_mnist()
```
